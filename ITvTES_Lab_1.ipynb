# 1 Задание
Составить файл-функцию для вычисления кусочнозаданной функции:

1) для аргумента х, вводимого с клавиатуры, вычислить значения у;

2) для массива х вычислить значения у и вывести график.

image.png

````python
import matplotlib.pyplot as plt
import numpy as np

# Функция
def f(x):
  # np.atleast_1d(x) позволяет принимать и числа, и массивы
  x = np.atleast_1d(x)
  # у numpy есть np.where, которая рабоает быстрее, можно через for, но дольше
  y = np.where(x < 2, np.sqrt(4*(x**2)+1), np.exp(x)-1)

  return y
````

## 1.1 Ввод с клавиатуры

````python
# Условия с try-except
try:
  x_val = float(input("Введите значение x: ")) # Только числа в диапазоне [0; 2)
  y_val = f(x_val)
  if not(x_val >= 2 or x_val < 0):
    print(f"При x = {x_val}, f(x) = {y_val}")
  else:
    print("Не соблюдены условия: x = [0; 2)")
except ValueError:
  print("Произошла ошибка")
````

## 1.2 Работа с массивом и график

````python
# От нуля до двух 200 точек
x_array = np.linspace(0, 2, 200)
y_array = f(x_array)

# Настройки для графика
plt.figure(figsize = (10, 6))
plt.plot(x_array, y_array, label = "f(x)", color = "red", linewidth = 2)
plt.title("График кусочно-заданной функции")
plt.xlabel("x")
plt.ylabel("f(x)")

# Вывод графика
plt.show()
````

# 2 Задание
Составить программу для таблицы аргументов и значений функции Z на отрезке [$х_0$, $x_k$] с шагом Δх, записать таблицу в файл. На Python можно использовать numpy или pandas.

image.png

````python
import pandas as pd

# Параметры по условиям
x0 = 0
xk = 4.5
step = 0.5

# Массив (xk+step, чтобы захватить правую границу)
x_values = np.arange(x0, xk+step, step)

# Значение Z
z_values = 1 + 2 * x_values - x_values**2

# Создаем датафрейм (таблицу)
df = pd.DataFrame({
    'x': x_values,
    'Z': z_values
})

# Вывод таблицы
print("Таблица значений функции: ")
print(df.to_string(index=False))
````

## 2.1 Вывод второго задания в файле results.txt
image.png

````python
# Запись в текстовый файл
df.to_csv('results.txt', sep = '\t', index=False)
print("Таблица успешно сохранена в файл 'results.txt'")
````

Данные в строке разделяются пробелами. Если студент не сдавал экзамен, то соответствующий элемент строки содержит значение 0. Пример строк файла:

| 12300012 | 4 | 3 | 5 | 5 |
| --- | --- | --- | --- | --- |
| 10492212 | 3 | 3 | 0 | 0 |

Подготовьте программу, с помощью которого можно установить:

1) сколько человек сдавали экзамен (по каждой дисциплине отдельно);

2) сколько оценок 5, 4, 3 и 2 было получено по каждой дисциплине отдельно;

3) сколько задолженностей имеет каждый студент;

4) если студент не имеет задолженностей, то какой у него средний балл за сессию;

5) сколько студентов на курсе не имеют долгов;

6) размер назначенной стипендии, если известно, что сдавшие все экзамены на «отлично» получают стипендию в размере A руб., а не имеющие ни одной тройки – B руб. (Размер стипендии задайте самостоятельно.)

Постройте диаграмму распределения количества студентов, получивших оценки 5, 4, 3 и 2 соответственно.

Подготовьте файл REZULT.DAT, каждая строка которого содержит следующую информацию: шифр студента, средний балл (0 – в случае имеющихся задолженностей) и размер назначенной стипендии (0 – если стипендия не назначена).

````python
# Создадим внутри программы файл SESSIA.DAT, чтобы каждый раз не загружать сюда этот датафрейм

data_str = """1 3 3 0 0
2 0 0 0 0
3 5 5 5 5
4 4 4 5 4
5 4 4 3 4
6 0 0 3 0
7 3 3 3 3
8 4 4 4 4
9 3 4 3 4
10 4 3 5 4
"""

with open('SESSIA.DAT', 'w') as f:
  f.write(data_str)

# Чтение файла в матрицу
def read_sessia(filename):
  return np.loadtxt(filename, dtype = int)

# Запись итоговой матрицы в файл
def save_result(filename, matrix):
  np.savetxt(filename, matrix, fmt = '%d %.1f %d', header = 'ID Avg_Score')

# Загрузка данных
data = read_sessia('SESSIA.DAT')
ids = data [:, 0]
grades = data[:, 1:]
num_exams = grades.shape[1]

# Стипендия
A = 5000
B = 3000

print('Отчет по сессии\n'+"="*20)

# Оценка > 0
students_pass_count = np.sum(grades > 0, axis = 0)
for i, count in enumerate(students_pass_count):
  print(f"Дисциплина {i+1}: сдавали {count-1} человек.")
````

## 4.1 Статистика

````python
# Статистика оценок по дисциплинам
all_counts = []
for i in range(num_exams):
  counts = {val: np.sum(grades[:, i] == val) for val in [5, 4, 3, 0]}
  all_counts.append(counts)
  print(f"Дисциплина {i+1}: 5: {counts[5]}, 4: {counts[4]}, 3: {counts[3]}, 0: {counts[0]}")

# Расчет по каждому студенту
rezult_matrix = []
no_debts_count = 0
for i in range(len(ids)):
    student_grades = grades[i]
    debts = np.sum(student_grades == 0) + np.sum(student_grades == 2)
    av_grade = np.mean(student_grades)
    avg_score = 0
    stipend = 0

    if debts == 0:
        no_debts_count += 1
        avg_score = np.mean(student_grades)

        # Расчет стипендии
        if np.all(student_grades == 5):
            stipend = A
        elif np.all(student_grades >= 4):
            stipend = B

    rezult_matrix.append([ids[i], avg_score, stipend])
    print(f"Студент {ids[i]}: долгов - {debts}, средний балл: {av_grade}, стипендия - {stipend}")

print(f"\nВсего студентов без долгов: {no_debts_count}")
````

## 4.2 Вывод графика

````python
save_result('REZULT.DAT', np.array(rezult_matrix))

labels = ['5', '4', '3', '0']
total_grades = [np.sum(grades == 5), np.sum(grades == 4), np.sum(grades == 3), np.sum(grades == 0)]

plt.figure(figsize=(8, 5))
plt.bar(labels, total_grades, color=['green', 'blue', 'orange', 'red'])
plt.title('Распределение оценок на курсе')
plt.xlabel('Оценка')
plt.ylabel('Количество')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
````

4. **В чем отличие цикла for от цикла while? Зачем операторы break и continue? Можно ли с помощью while и других операторов заменить for? А наоборот?**
* `for`: используется, когда известно количество итераций или нужно перебрать элементы коллекции.
* `while`: выполняется, пока условие истинно (используется, когда заранее неизвестно, когда условие выполнится).
* `break`: немедленно прерывает выполнение цикла и выходит из него.
* `continue`: пропускает остаток текущей итерации и переходит к следующей.
* `While` может полностью заменить for через ручной счетчик.
* И наоборот, через бесконечный цикл и `break`, но это менее удобно.

7. **Функции для сортировки элементов в массиве.**
* `np.sort(array)`: возвращает отсортированную копию массива.
* `array.sort()`: сортирует исходный массив на месте.
* `np.argsort(array)`: возвращает индексы, которые бы отсортировали массив.
* Для убывания: используется срез `np.sort(array)[::-1]`.
